```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/DataAnalyst.csv")
df.info()
df.head()
```

```python
df.drop(columns="Unnamed: 0",inplace=True)
display(df)
```

```python
def title_simplifier (title):
    if "business" in title.lower():
        return "Business Data Analyst"
    elif "data science" in title.lower() or 'data scientist' in title.lower():
        return "Data Science"
    elif "data warehouse" in title.lower():
        return "Data Warehouse Analyst"
    elif "data engineer" in title.lower():
        return "Data Engineer"
    elif "quality" in title.lower():
        return "Data Quality Analyst"
    elif "reporting" in title.lower():
        return "Reporting Data Analyst"
    elif "security" in title.lower():
        return "Data Security Analyst"
    elif "analyst" in title.lower():
        return "Data Analyst"
    else:
     return np.NaN

def seniority(title):
  if 'sr' in title.lower() or 'senior' in title.lower() or "II" in title.lower() or "2" in title.lower() :
        return 'Senior'
  elif  'lead' in title.lower() or 'principal' in title.lower() or "3" in title.lower():
    return "Lead"
  elif 'jr' in title.lower() or 'junior' in title.lower():
        return 'Junior'
  elif "Entry Level" in title.lower():
      return "Entry Level"
  else:
    return "Unknown / Non-Applicable"
```

```python
df["job_title"]=df["Job Title"].apply(title_simplifier)
display(df)
df["Seniority Level"]=df["Job Title"].apply(seniority)
display(df)
df.isnull().sum()
```

```python
df.drop(columns="Job Title",inplace=True)
display(df)
```

```python
import re
def salarySplitter(salary):
  salaryList=re.findall("\d+",salary)
  if len(salaryList)==2:
    min=int(salaryList[0])*1000
    max=int(salaryList[1])*1000
    avg=(min+max)/2
    return pd.Series([min,max,avg])
  else:
    return pd.Series([None,None,None])
salarySplitter("$78K-$104K (Glassdoor est.)")
```

```python
df["Salary Estimate"].apply(salarySplitter)
```

```python
df[["Minimum Salary","Maximum Salary","Average Salary"]]=df["Salary Estimate"].apply(salarySplitter)
```

```python
df.drop(columns="Salary Estimate",inplace=True)
display(df)
```

```python
#we have to replace all -1s with NaN, as some values are not logical compared to the domain scope
df.replace({-1,-1.0,"-1","-1.0"},np.NaN,inplace=True)
df.info()
display(df)
df.isnull().sum()
```

```python
#We can observe that the rating column has got its rating in the company name as well
#We just want to make sure that the null ratings have null ratings in the company name column too
#We'll see whenever the rating column was empty, the rating in the company name was empty or not
display(df[df["Rating"].isnull()][["Rating","Company Name"]])
#We made sure that both columns have no ratings
df["Rating"].value_counts()
#Filling null values in rating column with 0
df.fillna({"Rating":0},inplace=True)
display(df)
display(df.isnull().sum())
display(df["Rating"].value_counts())
```

```python
df.dropna(subset=["Company Name"],inplace=True)
display(df.isnull().sum())
```

```python
df["Headquarters"].fillna("Unknown",inplace=True)
df["Easy Apply"].fillna(False,inplace=True)
df.isnull().sum()
```

```python
#We can transform the competitors column into the number of competitors in order to avoid any redundancy
#df["Competitors"].fillna("Unknown",inplace=True)
def competitorsToNumber(comp):
  #A built-in function checks whether the value bypassed as a 1st parameter matches the second parameter type or not
  if isinstance(comp,str):
    return len(comp.split(","))
  else:
    return 0
df["Competitors"].apply(competitorsToNumber)
df["No. of Competitors"]=df["Competitors"].apply(competitorsToNumber)
display(df)
```

```python
#pd.options.display.max_colwidth=None
df[df["No. of Competitors"]==4][["Competitors"]]
```

```python
df[df["Rating"]==df["Rating"].max()]
```

```python
df.reset_index(inplace=True,drop=True)
#To revert the column width back to its normal form
pd.reset_option('display.max_colwidth')
display(df.groupby("Founded").agg({"Average Salary":"sum"}).sort_values(by="Average Salary",ascending=False).head(20))
```

```python
#lambda is a function definition blueprint which is used for short code
python_1=lambda x:1 if "python" in x.lower() else 0

df["Python yn"]=df["Job Description"].apply(python_1)
display(df)
```
